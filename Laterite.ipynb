```python
from openpyxl import Workbook
from openpyxl.utils.dataframe import dataframe_to_rows

# Load your eligible households dataset (all_eligible_yourname_yyyymmdd.csv)
eligible_dataset = pd.read_csv("/content/all_eligible_Enock Mwizerwa_20230918.csv")  # Replace with your actual file path

# Create a Pandas Excel writer using openpyxl as the engine
excel_writer = pd.ExcelWriter("eligible_households_per_district.xlsx", engine='openpyxl')

# Iterate through each district
for district in eligible_dataset['District of residence'].unique():
    # Filter eligible households for the current district
    district_df = eligible_dataset[eligible_dataset['District of residence'] == district]

    # Create a new Excel workbook
    workbook = Workbook()

    # Add the current district's data to the worksheet
    worksheet = workbook.active
    worksheet.title = "Eligible_Households"
    for row in dataframe_to_rows(district_df, index=False, header=True):
        worksheet.append(row)

    # Save the worksheet in the Excel writer
    excel_writer.book.add_worksheet(worksheet)

# Save the Excel file
excel_writer.save()
```

```python
from openpyxl import Workbook
from openpyxl.utils.dataframe import dataframe_to_rows

# Load your eligible households dataset (all_eligible_yourname_yyyymmdd.csv)
eligible_dataset = pd.read_csv("/content/all_eligible_Enock Mwizerwa_20230918.csv")  # Replace with your actual file path

# Create a Pandas Excel writer using openpyxl as the engine
excel_writer = pd.ExcelWriter("eligible_households_per_district.xlsx", engine='openpyxl')

# Create a new Excel workbook
workbook = Workbook()

# Iterate through each district
for district in eligible_dataset['District of residence'].unique():
    # Filter eligible households for the current district
    district_df = eligible_dataset[eligible_dataset['District of residence'] == district]

    # Add the current district's data to a new worksheet
    worksheet = workbook.create_sheet(title=district)
    for row in dataframe_to_rows(district_df, index=False, header=True):
        worksheet.append(row)

# Remove the default sheet created by openpyxl
default_sheet = workbook['Sheet']
workbook.remove(default_sheet)

# Save the Excel file
excel_writer.book = workbook
excel_writer.save()
```
